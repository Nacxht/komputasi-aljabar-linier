# Implementasi Algoritma Gram-Schmidt

Pada bagian ini, kita mengimplementasikan algoritma Gram-Schmidt yang merupakan metode fundamental untuk dekomposisi $QR$. Algoritma ini bekerja dengan mengubah sekumpulan vektor yang saling bebas menjadi sekumpulan vektor yang ortonormal. Proses ini sangat penting dalam faktorisasi $QR$ dimana matriks $A$ dipecah menjadi produk matriks ortogonal $Q$ dan matriks segitiga atas $R$.

```python
def gram_schmidt(A):
    A = np.array(A, dtype=float)

    n = A.shape[1]

    Q = np.zeros_like(A, dtype=float)
    R = np.zeros((n, n), dtype=float)

    for j in range(n):
        # Ambil kolom ke-j
        v = A[:, j]

        # Kurangi proyeksi terhadap q sebelumnya
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])

            v = v - R[i, j] * Q[:, i]

        # Normalisasi
        R[j, j] = np.linalg.norm(v)

        Q[:, j] = v / R[j, j]

    return Q, R
```

# Implementasi Iterasi QR

Pada bagian ini, kita mengimplementasikan algoritma iterasi $QR$ untuk mencari nilai eigen ($eigenvalue$) dari suatu matriks. Algoritma ini bekerja dengan melakukan faktorisasi $QR$ secara berulang pada matriks $A$. Pada setiap iterasi, matriks $A_k$ dipecah menjadi $Q_k$ dan $R_k$, kemudian matriks baru $A_{k+1}$ dibentuk dengan mengalikan $R_k$ dengan $Q_k$. Proses ini diulang hingga matriks konvergen ke bentuk segitiga atas yang elemen diagonalnya merupakan nilai eigen dari matriks asli.

```python
def qr_iteration(A, iterations=10):
    A_k = np.array(A, dtype=float)

    for k in range(iterations):
        print(f"\n=== Iterasi {k+1} ===")

        # Faktorisasi QR
        Q, R = gram_schmidt(A_k)

        print("\nQ:")
        print(Q)

        print("\nR:")
        print(R)

        # A_(k+1) = RQ
        A_k = R @ Q

        print("\nA baru:")
        print(A_k)

    return A_k
```

# Contoh Implementasi dan Hasil

Pada bagian ini, kita menerapkan algoritma iterasi $QR$ pada contoh matriks $2 \times 2$ sederhana. Matriks yang digunakan adalah $[[2, 1], [1, 2]]$ yang merupakan matriks simetris. Kita melakukan 10 iterasi untuk menunjukkan konvergensi algoritma. Hasil akhir yang diharapkan adalah matriks segitiga atas dengan nilai eigen pada diagonalnya. Untuk matriks ini, nilai eigen yang diharapkan adalah 3 dan 1.

```python
A = np.array([
    [2, 1],
    [1, 2]
], dtype=float)

hasil = qr_iteration(A, iterations=10)

print("Hasil Akhir:")
print(hasil)

print("\nPerkiraan Eigenvalue:")
print(np.diag(hasil))
```
